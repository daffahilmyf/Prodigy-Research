```python
import json

en = open('en-us.json', 'r', encoding='utf-8')
jp = open('ja-jp.json', 'r', encoding='utf-8')

data1 = json.load(en)
data2 = json.load(jp)
```

```python
# create a new dictionary to hold the merged data
merged_data = {}
merged_object_data = {}
# loop over the keys in the first file
for key in data1.keys():
    # skip over keys with array or object values
    if isinstance(data1[key], (list, dict)) or isinstance(data2.get(key), (list, dict)):
        merged_object_data[key] = (data1[key], data2[key])
        continue


    # if the key exists in the second file, merge the values into a tuple
    if key in data2:
        merged_data[key] = (data1[key], data2[key])

# print out the merged data
for key, values in merged_data.items():
    print(f"{key}; {values[0]}; {values[1]}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import json


# separate the text and label into separate lists
texts = [d['text'] for d in data]
labels = [d['label'] for d in data]

# encode the labels as numeric values
le = LabelEncoder()
labels = le.fit_transform(labels)

# perform feature extraction using the CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts).toarray()

# create and train the logistic regression model
model = LogisticRegression()
model.fit(X, labels)

```
